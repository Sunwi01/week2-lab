## 2. NumPy ‚Äì From Lists to Arrays

AI models require **fast numerical computation**.
NumPy converts Python lists into efficient numerical arrays.

```python
import numpy as np

areas_np = np.array([50, 60, 80, 100, 120])
prices_np = np.array([25000, 30000, 40000, 50000, 60000])

print("Areas:", areas_np)
print("Average area:", np.mean(areas_np))
```

### ‚úèÔ∏è Exercise 2: NumPy Statistics

Using NumPy, compute:
- average price

```python
import numpy as np
# TODO: convert lists to numpy arrays
areas_np = np.array([50, 60, 80, 100, 120])
prices_np = np.array([25000, 30000, 40000, 50000, 60000])

# TODO: statistics
print("Average price:", np.mean(prices_np))
print("Max price:", np.max(prices_np))
print("Min area:", np.min(areas_np))
```

## 3. Matplotlib ‚Äì Visualizing House Prices

Before training AI models, we **look at the data**.
Scatter plots show relationships between variables.

```python
import matplotlib.pyplot as plt

plt.scatter(areas_np, prices_np)
plt.xlabel("House Area (m¬≤)")
plt.ylabel("House Price ($)")
plt.title("House Price vs Area")
plt.show()
```

### ‚úèÔ∏è Exercise 3: Scatter Plot

Plot your own house data using a scatter plot.

```python
# TODO: scatter plot
import matplotlib.pyplot as plt
import numpy as np

areas_np = np.array([50, 60, 80, 100, 120])
prices_np = np.array([25000, 30000, 40000, 50000, 60000])

plt.scatter( areas_np, prices_np)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Relation between area and price")
plt.show()
```

## 4. Machine Learning ‚Äì House Price Prediction

Now we let the computer **learn the relationship** between area and price.

This is a **regression problem**.

```python
from sklearn.linear_model import LinearRegression

# Input must be 2D for scikit‚Äëlearn
X = areas_np.reshape(-1, 1)
y = prices_np

model = LinearRegression()
model.fit(X, y)

print("Slope (price per m¬≤):", model.coef_[0])
print("Base price:", model.intercept_)
```

```python
# Predict price for a new house
from sklearn.linear_model import LinearRegression

new_area = 90
predicted_price = model.predict([[new_area]])


print("Predicted price for", new_area, "m¬≤:", predicted_price[0])
```
